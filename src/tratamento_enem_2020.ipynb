## 2. Carregamento dos dados

```python
dtype_map = {
    'NU_INSCRICAO': str,
    'NU_ANO': 'int16',
    'TP_FAIXA_ETARIA': 'Int8', 'TP_SEXO': str, 'TP_ESTADO_CIVIL': 'Int8',
    'TP_COR_RACA': 'Int8', 'TP_NACIONALIDADE': 'Int8', 'TP_ST_CONCLUSAO': 'Int8',
    'TP_ANO_CONCLUIU': 'Int8', 'TP_ESCOLA': 'Int8', 'TP_ENSINO': 'Int8',
    'IN_TREINEIRO': 'Int8',
    'CO_MUNICIPIO_ESC': 'Int32', 'NO_MUNICIPIO_ESC': str,
    'CO_UF_ESC': 'Int8', 'SG_UF_ESC': str,
    'TP_DEPENDENCIA_ADM_ESC': 'Int8', 'TP_LOCALIZACAO_ESC': 'Int8', 'TP_SIT_FUNC_ESC': 'Int8',
    'CO_MUNICIPIO_PROVA': 'Int32', 'NO_MUNICIPIO_PROVA': str,
    'CO_UF_PROVA': 'Int8', 'SG_UF_PROVA': str,
    'TP_PRESENCA_CN': 'Int8', 'TP_PRESENCA_CH': 'Int8',
    'TP_PRESENCA_LC': 'Int8', 'TP_PRESENCA_MT': 'Int8',
    'NU_NOTA_CN': 'float32', 'NU_NOTA_CH': 'float32',
    'NU_NOTA_LC': 'float32', 'NU_NOTA_MT': 'float32',
    'TP_LINGUA': 'Int8', 'TP_STATUS_REDACAO': 'Int8',
    'NU_NOTA_COMP1': 'Int16', 'NU_NOTA_COMP2': 'Int16',
    'NU_NOTA_COMP3': 'Int16', 'NU_NOTA_COMP4': 'Int16',
    'NU_NOTA_COMP5': 'Int16', 'NU_NOTA_REDACAO': 'float32',
}

df = pd.read_csv(
    DADOS_PATH,
    sep=';',
    encoding='latin-1',
    usecols=list(rename_map.keys()),
    dtype=dtype_map,
)
print(f'Shape original: {df.shape}')
print(f'Colunas: {df.columns.tolist()}')
```

## 3. Renomeação das colunas

```python
df = df.rename(columns=rename_map)
print(f'Shape após renomeação: {df.shape}')
print(f'Novas colunas: {df.columns.tolist()}')
```

## 4. Exportação dos dados tratados

```python
PROCESSED_PATH = r'..\data\processed\enem_2020_tratado.parquet'
df.to_parquet(PROCESSED_PATH, index=False)
print(f'Dados exportados para {PROCESSED_PATH}')
print(f'Shape: {df.shape}')
```
